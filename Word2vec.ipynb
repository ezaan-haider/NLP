```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []
for filename in os.listdir("/hpbooks"):
  f = open(os.path.join("/hpbooks", filename))
  corpus = f.read()
  raw_sent = sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
story = []
folder = "/hpbooks"

for filename in sorted(os.listdir(folder)):
    print("Reading:", filename)
    path = os.path.join(folder, filename)
    with open(path, encoding='utf-8') as f:
        corpus = f.read()
        raw_sent = sent_tokenize(corpus)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
print("First sentence tokens:", story[0])
print("Raw text:", ' '.join(story[0]))
```

```python
model = gensim.models.Word2Vec(
    window=1,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar("harry")
```

```python
model.wv.most_similar("snape")
```

```python
model.wv.doesnt_match(["harry", "ron", "hermione", "malfoy"])
```
